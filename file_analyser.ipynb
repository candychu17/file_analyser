# File Extension Analyser :)

1) Import Python packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import math
from itertools import product
import datetime as dt
```

2) Load Teracopy CSV into a DataFrame 

```python

df = pd.read_csv ('data/Teracopy.csv')
df.head()
```

add size columns for MB and GB

```python
df['Size(MB)'] = df['Size'].div(1000000)
df['Size(GB)'] = df['Size'].div(1000000000)

df.head()
```

3) Add new column 'Extension' while copying over the full path values

```python
#df['Extension']=0
#df.head()
df['Extension'] = df['Full path']
df.head()
```

4) Iterate through rows of Extension column to retrieve just the extension name

```python
for index in range(len(df)):
    path = df.loc[index,"Extension"]

    root_ext = os.path.splitext(path)

    clean_ext = root_ext[1].replace('.','')

    df.loc[index,"Extension"] = clean_ext

df.head()
```

5) Add new Column 'Folder' while copying over full path values

```python
df['Folder'] = df['Filename']
```

6) Iterate through rows of Folder column to retrieve just the folder path

```python
for index in range(len(df)):
    path = df.loc[index,"Folder"]

    root_ext = os.path.splitext(path)

    folder_path = root_ext[0]

    #if (df.loc['Extension'!=""]):

    df.loc[index,"Folder"] = folder_path 

df.head()
```

7) Indicate if it is a folder

```python
for index in range(len(df)):
    if df.loc[index,'Extension']=='':
        df.loc[index,"Folder"] = df.loc[index,'Filename']
    else:
        df.loc[index,"Folder"] = ''

df.head()
```

8) Pull in Reference CSV

```python
dfr = pd.read_csv ('data/Reference.csv')
dfr.head()
```

9) Inner left join both dataframes based on Extension Column

```python
mergeddf = pd.merge(df,dfr, on ="Extension", how = "left")
del mergeddf['SEQ']
del mergeddf['Description']
mergeddf.head()
```

10) Fill all NaN values with ''. For any files without matching matching reference extension; classified under 'Uncategorised'

```python
mergeddf.fillna('', inplace=True)

for index in range(len(mergeddf)):
    if mergeddf.loc[index,'Extension']!='':
        if mergeddf.loc[index,'Category'] == '':
            mergeddf.loc[index,'Category'] = 'Uncategorised'
    else:
        mergeddf.loc[index,'Category'] = 'Folder'

mergeddf.head()
```

```python
mergeddf.dtypes
```

```python
mergeddf['Created']=pd.to_datetime(mergeddf.Created)
mergeddf['Updated']=pd.to_datetime(mergeddf.Updated)



#convert date format
mergeddf['Created']=mergeddf.Created.dt.strftime("%x")
mergeddf['Updated']=mergeddf.Updated.dt.strftime('%x')

#mergeddf['Created']=pd.to_datetime(mergeddf.Created)
#mergeddf['Updated']=pd.to_datetime(mergeddf.Updated)

mergeddf.dtypes

#mergeddf
```

Insert the Date Range of which user is interested here

```python

mergeddf=mergeddf.loc[(mergeddf["Created"]>="06/08/22") & (mergeddf["Updated"]<="06/12/22")]

mergeddf
```

11) User can see the list of categories available to choose from. 

```python
print(mergeddf['Category'].unique())
```

12) User is able to input criteria selection in --> criteria = {"Category": ["x","y","z"]}

```python
criteria = {"Category": ["Graphics Files","Uncategorised","Folder"]}
```

13) User selected criteria will be filtered out to produce final dataframe

```python
def filterby_criteria(df,criteria):
    k,v = criteria.keys(),criteria.values()
    all_matches = (pd.DataFrame(product(*v))).set_axis(list(k),axis='columns',inplace=False)
    return df.merge(all_matches)

final_df = filterby_criteria(mergeddf,criteria)

final_df.head()
```

14) Fill up all '' with NaN for reporting purposes

```python
for index in range(len(final_df)):
    if final_df.loc[index,'Extension']=='':
        final_df.loc[index,'Extension']=np.nan
    if final_df.loc[index,'Folder']=='':
        final_df.loc[index,'Folder']=np.nan

final_df.head()
```

15) Shows the size of each type of extensions in each category

```python
each_category = final_df.groupby(['Category','Extension','Size'])

each_category=each_category.sum()

del each_category['Index']

each_category
```

16) Shows the total number of files in each category

```python
total_file=final_df.groupby(['Category']).count()

total_file['Total_Files']=total_file['Full path']

total_file=total_file[['Total_Files']]

total_file
```

17) Shows the total size of each category

```python
total_size=final_df.groupby(['Category']).sum(['Size'])

total_size=total_size[['Size']]

total_size['Total_Size']=total_size['Size']

total_size['Total_Size(MB)'] = total_size['Size'].div(1000000)

total_size['Total_Size(GB)'] = total_size['Size'].div(1000000000)

del total_size['Size']
del total_size['Total_Size']

total_size
```

```python
reportdf = pd.merge(total_file,total_size, on ="Category", how = "left")

reportdf
```
